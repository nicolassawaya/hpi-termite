```python

# pathlist = ['hsa04623']

# for pathid in pathlist:

# Path
pathid = 'hsa04623'
print(pathid)
print(f"*** {pathnames[pathid]} ***")

# Read file
fpath = dirpath+f"{pathid}.json"
G = termite.pathways.load_pathway_graph(fpath)
print(G)

# Remove single-node components
edge_list = [edge for edge in G.edges()]
nodes_from_edges = set(itertools.chain.from_iterable(edge_list))
Gsub = G.subgraph(nodes_from_edges)
print(Gsub)

# 
labels = nx.get_node_attributes(Gsub, 'names') 
#     print(labels)


xnode = '98'
sublabels = {xnode:labels['98'], }



nx.draw_kamada_kawai(Gsub,node_size=60,alpha=0.5) #, labels=sublabels)#,node_color=colors)
plt.show()




```

```python
print("*** Human-viral PPi found ***")

colors = dict( zip( labels.keys(),['blue']*len(labels.keys()) ) )
#     print(colors)
colors[xnode] = 'red'
# for nodeid in children:
#     colors[nodeid] = 'darkorange'
# for nodeid in grandchildren:
#     colors[nodeid] = 'orange'
colors = list(colors.values())


#     nx.draw_kamada_kawai(Gsub,node_size=60,alpha=0.5, with_labels=True)
nx.draw_kamada_kawai(Gsub,node_size=60,alpha=0.5, labels=sublabels,node_color=colors)
plt.show()

```

```python
print("*** Plotting downstream effects ***")



print(f"Predecessors of node {xnode}")
print( list(Gsub.predecessors(xnode)) )

children = list(Gsub.successors(xnode))
grandchildren = set()
for child in children:
    grandchildren.update(Gsub.successors(child))
print(f"Successors of node {xnode}")
print(children)
print(grandchildren)

colors = dict( zip( labels.keys(),['blue']*len(labels.keys()) ) )
#     print(colors)
colors[xnode] = 'red'
for nodeid in children:
    colors[nodeid] = 'darkorange'
for nodeid in grandchildren:
    colors[nodeid] = 'orange'
colors = list(colors.values())
#     print(colors)



#     nx.draw_kamada_kawai(Gsub,node_size=60,alpha=0.5, with_labels=True)
nx.draw_kamada_kawai(Gsub,node_size=60,alpha=0.5, labels=sublabels,node_color=colors)
plt.show()
```
